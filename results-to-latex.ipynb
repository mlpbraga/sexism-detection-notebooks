```python
mean_results = report
```

```python
def map_key_to_latex(key):
    if 'B_SVM' in key:
        return 'Baseline SVM'
    if 'B_LR' in key:
        return 'Baseline LR'
    if 'TFus' in key:
        return '$TF_{us}$'
    if 'TFun' in key:
        return '$TF_{un}$'
    if 'TFbs' in key:
        return '$TF_{bs}$'
    if 'TFbn' in key:
        return '$TF_{bn}$'
    if 'TFuQ' in key:
        return '$TF_{us}$ + $TF_{un}$ + $Q_{l}$ + $Q_{s}$ + $Q_{c}$ + $Q_{p}$'
    if 'TFb' in key:
        return '$TF_{bs}$ + $TF_{bn}$'
    if 'TFu' in key:
        return '$TF_{us}$ + $TF_{un}$'
    if 'TFpos' in key:
        return '$TF_{pos}$'
    if 'CQWQ' in key:
        return '$Q_{c}$ + $Q_{p}$'
    if 'LD' in key:
        return '$Q_{l}$ + $Q_{d}$'
    if 'L' in key:
        return '$Q_{l}$'
    if 'D' in key:
        return '$Q_{d}$'
    if 'WQ' in key:
        return '$Q_{p}$'
    if 'CQ' in key:
        return '$Q_{c}$'
    if 'TF' in key:
        return '$TF_{us}$ + $TF_{un}$ + $TF_{bs}$ + $TF_{bn}$'
    if 'Q' in key:
        return '$Q_{l}$ + $Q_{d}$ + $Q_{c}$ + $Q_{p}$'
    else:
        return key

def print_to_latex(report):
    return ('&{%.5f}&{%.5f}&{%.5f}&{%.5f}&{%.5f}&{%.5f} \\\\' % ((report["precision"]["1"]),
                                                                (report["precision"]["0"]),
                                                                (report["recall"]["1"]),
                                                                (report["recall"]["0"]),
                                                                (report["f1"]["1"]),
                                                                (report["f1"]["0"])))

columns = ['$TF_{us}$',
           '$TF_{un}$',
           '$TF_{bs}$',
           '$TF_{bn}$',
           '$Q_{l}$',
           '$Q_{d}$',
           '$Q_{c}$',
           '$Q_{p}$',
           '$Q_{c}$ + $Q_{p}$',
           '$Q_{l}$ + $Q_{d}$',
           '$Q_{l}$ + $Q_{s}$ + $Q_{c}$ + $Q_{p}$',
           '$TF_{pos}$',
           '$TF_{us}$ + $TF_{un}$',
           '$TF_{bs}$ + $TF_{bn}$',
           '$TF_{us}$ + $TF_{un}$ + $TF_{bs}$ + $TF_{bn}$',
           '$TF_{us}$ + $TF_{un}$ + $Q_{l}$ + $Q_{s}$ + $Q_{c}$ + $Q_{p}$',]

print('SVM')
result_latex = []
for key in mean_results.keys():
    if 'SVM' in key:
        print(map_key_to_latex(key),print_to_latex(mean_results[key]))

print('KNN')
result_latex = []
for key in mean_results.keys():
    if 'KNN' in key:
        print(map_key_to_latex(key),print_to_latex(mean_results[key]))

print('RFC')   
result_latex = []
for key in mean_results.keys():
    if 'RFC' in key:
        print(map_key_to_latex(key),print_to_latex(mean_results[key]))
```
